### Plotting the bar chart where the Variable 1 is RESPONSES and Variable 2 is AGE

```python
data_filtered = data[(data['If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?'] != '-') & (data['Age'] != '-')]

pivot_table = data_filtered.pivot_table(index='Age', 
                                        columns='If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?', 
                                        aggfunc='size', 
                                        fill_value=0)

percentages = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

colors = ['#d9d696', '#d996b8', '#a096d9', '#96c1d9']

fig, ax = plt.subplots(figsize=(8, 4))

bars = percentages.plot(kind='barh', stacked=True, ax=ax, color=colors)
plt.ylabel('Age', color='#38383b')

legend = plt.legend(bbox_to_anchor=(0.5, 0.95), loc='upper center', ncol=4, fontsize='small', bbox_transform=plt.gcf().transFigure)
plt.setp(legend.get_texts(), color='#38383b')

ax.set_xticklabels([], color='#38383b')
ax.tick_params(axis='x', which='both', length=0)

for bar in bars.containers:
    for rect in bar:
        width = rect.get_width()
        label_x = rect.get_x() + rect.get_width() / 2
        label_y = rect.get_y() + rect.get_height() / 2
        ax.annotate(f'{int(width)}%', 
                    xy=(label_x, label_y),
                    xytext=(0, 0),
                    textcoords="offset points",
                    ha='center', va='center',
                    color='#38383b')

ax.spines['bottom'].set_color('#38383b')
ax.spines['left'].set_color('#38383b') 
ax.spines['top'].set_color('#38383b') 
ax.spines['right'].set_color('#38383b')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# Createing a grid of plots to show 2 plots and a description

```python
data_filtered = data[(data['If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?'] != '-') & (data['Age'] != '-')]

# Create pivot table for percentages
pivot_table_percentage = data_filtered.pivot_table(index='Age', columns='If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?', aggfunc='size', fill_value=0)
percentages = pivot_table_percentage.div(pivot_table_percentage.sum(axis=1), axis=0) * 100

pivot_table_counts = data_filtered.pivot_table(index='Age', columns='If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?', aggfunc='size', fill_value=0)

colors = ['#d9d696', '#d996b8', '#a096d9', '#96c1d9']

fig, axs = plt.subplots(1, 2, figsize=(18, 6))

bars1 = percentages.plot(kind='barh', stacked=True, ax=axs[0], color=colors)
axs[0].set_title('Percentage of Responses by Age')
axs[0].set_xlabel('Percentage')
axs[0].set_ylabel('')

bars2 = pivot_table_counts.plot(kind='barh', stacked=True, ax=axs[1], color=colors)
axs[1].set_title('Counts of Responses by Age')
axs[1].set_xlabel('Count')
axs[1].set_ylabel('')

axs[0].legend().set_visible(False)
axs[1].legend().set_visible(False)

# Adding legend above the grap
fig.legend(labels=pivot_table_percentage.columns, loc='upper center', ncol=len(pivot_table_percentage.columns), bbox_to_anchor=(0.5, 1.0))

# Adding description below the graphs
description = """
- Chart 1 shows the percentage of responses by age, grouped by the likelihood of checking weather on a smartwatch. Among younger age groups, particularly those under 30, there's a higher proportion of individuals likely to use a smartwatch for weather updates. As age increases, the likelihood of using a smartwatch for weather checking decreases gradually, with older age groups showing lower percentages in the "very likely" category. This suggests a generational trend where younger individuals are more inclined towards adopting smartwatch technology for everyday tasks, including checking the weather.

- Chart 2 displays the counts of responses by age, similarly grouped. We observe variations in weather-checking behavior across different age groups. While younger age groups have a higher count of respondents overall, the proportion of individuals likely to use a smartwatch for weather checking varies within each age group. Despite the higher count of respondents in younger age groups, the absolute number of individuals likely to use a smartwatch for weather updates might be similar or even higher in some older age groups due to variations in likelihood within each age category.
This highlights the importance of considering both percentages and counts when analyzing the data, as it offers a more comprehensive view of the distribution of responses across different demographics.
"""
fig.text(0.5, -0.3, description, horizontalalignment='center', fontsize=12, wrap=True, bbox=dict(facecolor='white', alpha=0.5))

plt.subplots_adjust(left=0.1, right=0.9)  # Adjust left and right margins

plt.show()
```
